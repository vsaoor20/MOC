```python
#Method of characteristics

import math
def Prandtl_meyer(M,gamma):
    A=math.sqrt((gamma+1)/(gamma-1))
    B=math.sqrt((gamma-1)/(gamma+1))
    C=math.sqrt(math.pow(M,2)-1)    
    mu=(A*math.degrees(math.atan(B*C)))-math.degrees(math.atan(C))
    return mu

def inv_pm(nu):
    nu=nu*math.pi/180;
    A=1.3604;
    B=0.0962;
    C=-0.5127;
    D=-0.6722;
    E=-0.3278;
    nu_0=0.5*math.pi*(math.sqrt(6)-1)
    y=math.pow((nu/nu_0),(2/3))
    m=(1+(A*y)+(B*math.pow(y,2))+(C*math.pow(y,3)))/(1+(D*y)+(E*math.pow(y,2)))
    return round(m,2)

def mach_angle(mach):
    mu=math.asin(1/mach)*(180/math.pi)
    return round(mu,2)

exit_mach=float(input("Enter exit mach number: "));
exit_nu=round(Prandtl_meyer(exit_mach,1.4),2);
theta_max=round(exit_nu/2,3);
div=int(input("Enter the number of divisions: "));
#counting the number of points
count=0;

#defining thetha_initial and delta_theta

theta_initial=theta_max-math.floor(theta_max)
delta_theta=math.floor(theta_max)/(div-1)

print("Theta initial: {}".format(theta_initial))
print("Delta theta: {}".format(delta_theta))

#exit nu
print("Exit nu: {}".format(exit_nu))
print("Theta_max: {}".format(theta_max))

#counting the number of poins

for i in range(1,div+1,1):
    count=count+i+1;
print("The total number of points: "+str(count))

#Creating a 2D Array format Point(Array_index),K+,K-,Theta, Nu, Mach and Mu
#Rows=35 Cols=6

data=[0]*(count+1)
for i in range(count+1): data[i]=[0]*7

#Points of first left running characteristics

pt_count=0;

#we are putting values from 0

for i in range(1,div+1):
    pt_count=pt_count+1;
    data[i][0]=pt_count;
    data[i][2]=0;
    if(pt_count==1):
        data[i][3]=round(theta_initial,3)
    else:
        data[i][3]=round(data[i-1][3]+delta_theta,3)
    data[i][4]=data[i][3];
    data[i][1]=2*data[i][4];
    data[i][5]=inv_pm(data[i][4])
    data[i][6]=mach_angle(data[i][5])
    
#defining first wall point
pt_count=pt_count+1
data[div+1][0]=pt_count
for j in range(1,7):
    data[div+1][j]=data[div][j];

#defining other points 

for k in range(1,div):
    pt_count=pt_count+1;
    take=k+1;
    data[pt_count][0]=pt_count;
    data[pt_count][3]=0;
    data[pt_count][1]=data[take][1];
    data[pt_count][4]=data[pt_count][1];
    data[pt_count][2]=data[pt_count][3]-data[pt_count][4];
    data[pt_count][5]=inv_pm(data[pt_count][4])
    data[pt_count][6]=mach_angle(data[pt_count][5])  
    constant=data[pt_count][3];
    pt_count=pt_count+1
    start=pt_count
    end=pt_count+div-k-1
    for i in range(pt_count,pt_count+div-k-1):
        
        data[i][0]=pt_count;
        take=take+1
        data[i][2]=constant;
        data[i][1]=data[take][1];
        data[i][3]=0.5*(data[i][1]+data[i][2])
        data[i][4]=0.5*(data[i][1]-data[i][2])
        data[i][5]=inv_pm(data[i][4])
        data[i][6]=mach_angle(data[i][5])
        pt_count=pt_count+1
    
        
    data[pt_count][0]=pt_count;
    for j in range(1,7):
        data[pt_count][j]=data[pt_count-1][j]

#O/P Data points        
print ("{:<8} {:<15} {:<10}{:<8} {:<15} {:<10} {:<10}".format('Point','K-','K+','Theta','Nu','M','Mu'))
#for i in range(1,count+1):
    #for j in range(7):
        #print(round(data[i][j],2),end=" ")
    #print("\n")
for i in range(1,count+1):
    p,k1,k2,t,n,m,mu = data[i]
    print("{:<8} {:<15} {:<10}{:<8} {:<15} {:<10} {:<10}".format(p,k1,k2,t,n,m,mu))

        
```
